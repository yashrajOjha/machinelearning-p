# Regression

Regression analysis is a statistical method that helps us to analyse and understand the relationship between two or more variables of interest. 

- **Dependent Variable**
    It is something we are trying to explain, like how will a person's salary change by experience, and so salary is dependent on years of experience.
    Something that we are trying to understand how it varies with another variable.

- **Independent Variable**
    These are factors that influence the analysis or target variable and provide us with information regarding the relationship of the variables with the target variable.
    
### Simple Linear Regression

Let assume a relation between Salary and Experience, a simple formula will be 
    
    Salary = constant + (co.eff * Experience)
    
Salary is a dependent variable, constant might be like the basic salary, that if if we draw a line chart, where the line starts from will be the constant.

    Salary = 30k + co.eff * Experience (30k might be the starting point)
   So when experience is 0, the salary is just 30k, with experience it increase.
   
Co.eff is the slope of the line, if it is less then, a lower salary gain with experience, it the slope is more then it is a higher salary gain.



---

**Ordinary Least Square Method**

Simple Regression is based on the slope/trend line. We first locate the actual salary the employee earns and then we see the salary the model has predicted for that experience using the trend line. 



The green line is the difference between what he is earning and what he should be earning. ]

To get the best trend line, we take all the green line distances and square them and then sum them, once we have sum of the squares we need to find the minimum among the sum of all the other lines. 

The model draws multiple lines and using the above method to find the line which is the best, where the difference is minimum.

---
So to put this all into simpler terms, this is the branch of machine learning that aims to predict some continuous real numbers, for example, a salary or a temperature or any kind of continuous numerical value.

```python
#importing libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing data set
dataset = pd.read_csv('Salary_Data.csv')
```

```python
x = dataset.iloc[:,:-1].values #all the rows and all the columns except last one
y= dataset.iloc[:,-1].values #all the rows and the last column
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
lrmodel = LinearRegression() # object of Linear Regression class
lrmodel.fit(x_train,y_train) #training the regression model
#the format of fit is training sets
```

```python
predicted_salaries = lrmodel.predict(x_test)
```

```python
plt.scatter(x_train,y_train, color='r',s=10)
plt.plot(x_train,lrmodel.predict(x_train),color='b')
plt.title("Salary vs Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

```python
plt.scatter(x_test,y_test, color='r',s=10)
plt.plot(x_train,lrmodel.predict(x_train),color='b')
plt.title("Salary vs Experience (Test)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
```

#### How do I use my simple linear regression model to make a single prediction, for example, to predict the salary of an employee with 12 years of experience?

```python
print(lrmodel.predict([[12]]))
```

**Important note:** Notice that the value of the feature (12 years) was input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting 12 into a double pair of square brackets makes the input exactly a 2D array. Simply put:

$12 \rightarrow \textrm{scalar}$

$[12] \rightarrow \textrm{1D array}$

$[[12]] \rightarrow \textrm{2D array}$

#### How do I get the final regression equation y = b0 + b1 x with the final values of the coefficients b0 and b1?

```python
print(lrmodel.intercept_)
print(lrmodel.coef_)
```

To get these coefficients we called the "coef_" and "intercept_" attributes from our regressor object.
