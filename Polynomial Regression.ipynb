# Polynomial Regression

It is as simple as Multiple Linear Regression but the powers of the variables can vary.
In linear regression we have a simple **y=mx+c**, but if the points are arranged in a parabolic way in a graph we need to have a polynomial equation.

    y = b + b1x1 + b2x1 ^2 + ... +bnxn ^n
    
### What is it with linear models?

When we say a model is linear we talk about the coefficients and not the x variable, we know the x variable, what we dont know are those coefficients. A model is linear if, **the dependent variable can be written as linear combination of the independent variable**.

A model is non linear when the coeff are written as a factor, for example:
    
     y = b + b1/b2 x1 + b2/b3b4 x2
     
Then this would become a non linear equation.

**Program to predict salary of a particular position.**

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lrmodel = LinearRegression()
lrmodel.fit(x,y)
```

```python
#making a power matrix of the features
from sklearn.preprocessing import PolynomialFeatures
pmodel = PolynomialFeatures(degree=4)
#build a polynomial model where max power is 2
#tranforming the matrix from basic to x,x2
x_poly = pmodel.fit_transform(x)
#after transforming to the polynomial matrix
#we feed the matrix to our linear regression model
lrmodel2 = LinearRegression()
lrmodel2.fit(x_poly,y)
```

#### Visualising the Linear Regression Model

```python
plt.scatter(x,y,c='red')
plt.plot(x,lrmodel.predict(x),c='blue')
plt.title("Using Simple Linear Regression on Non Linear Data",pad=15)
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

We can notice that the simple linear regression model isn't fitting the data well.

#### Visualising the Polynomial Regression Model

```python
plt.scatter(x,y,c='red')
plt.plot(x,lrmodel2.predict(x_poly),c='blue')
plt.title("Using Polynomial Regression on Non Linear Data (degree 2)",pad=15)
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x,y,c='red')
plt.plot(x,lrmodel2.predict(x_poly),c='blue')
plt.title("Using Polynomial Regression on Non Linear Data (degree 4)",pad=15)
plt.xlabel("Position")
plt.ylabel("Salary")
plt.show()
```

```python
#person asked for 160k salary for his position
#so lets see what our linear regression model suggests
lrmodel.predict([[6.5]])
```

```python
#so lets see what our polynomial regression model suggests
lrmodel2.predict(pmodel.fit_transform([[6.5]]))
```
