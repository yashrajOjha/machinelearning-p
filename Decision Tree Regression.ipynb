# Decision Tree Regression

CART is Classification and Regression trees.

Let's take two independent variables x1 and x2, after we run the DT algorithm, our scatter plot will be split into sections.


How and where these splits are happening is based on the algorithm.

Lets take the first split at x2<20 points, split 2 happens at x2=170, split 3 happens at x2<200, split 4 happens at x1<40. 

Lets take our observation is  x1=30 and x2=50.

After looking at the graph we can conclude that our point lies with **split 1 and split 4 (considering x1 value)** and it also lies within **split 2 (considering x2 value)**.

We need to find value of dependant variable 'Y', considering every split has an average value of the leafs(found using the points), y will be assigned a value of the average of the leaf,that the x1=30 and x2=50 lies in.



# Code 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

**Decision trees and ensemble methods do not require feature scaling to be performed as they are not sensitive to the the variance in the data.**

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state=0)
model.fit(x,y)
```

Decision tree do not guarantee the same solution globally. There will be variations in the tree structure each time you build a model. Passing a specific seed to random_state ensures the same result is generated each time you build the model.

```python
#predicting salary for 6.5 years of experience 
model.predict([[6.5]])
```

The decision tree regression model is not the best model to use on a single feature data set. It is more adapted to datasets with many features, that is a high dimensional dataset.

Also the it is advisable to use Decision Tree when the data is more non-parametric in nature

```python
x_grid=np.arange(min(x),max(x),0.1)
x_grid=x_grid.reshape((len(x_grid),1))
plt.scatter(x,y,color='red')
plt.plot(x_grid,model.predict(x_grid),color='blue')
plt.title('Decision Tree Model')
plt.xlabel('Position')
plt.ylabel('Salary')
```
